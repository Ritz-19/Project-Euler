# 4. A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 Ã— 99. Find the largest palindrome made from the product of two 3-digit numbers.

```python
#Fucntion to check if number is palindrome

def palin_checker(num):
  list1 = []
  listr = []
  copy = num
  s = str(num)
  list1 = list(s)

  for i in range(0,len(list1)):
    c = int(copy%10)
    listr.append(str(c))
    copy = int(copy/10)

  if list1 == listr:
    return True
```

```python
temp = 0
n = 3

if n == 2:
  l = range(60,100)
elif n == 3:
  l = range(600,1000)

for i in l:
  for j in l:
    prod = i*j
    if palin_checker(prod) == True and prod>temp:
      temp = prod
 
print("The largest palindrome is: {}".format(temp))
```

# 5. 2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder. What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?

```python
def find_prime(num):
  # function to find all prime numbers upto a certain number (num)
  list1 = [2]
 
  for j in range(3,num+1,2): #prime numbers are odd, except 2
    count = 0 #counter to check for factors
 
    for k in range(2,j+1):
      if j%k == 0:
        count += 1
      
    if count == 1:
      list1.append(j)
 
  return list1
 
def find_exp(num, p):
  # function to check the highest exponents of the primes

  list2 = []
 
  for i in range(0,len(p)):
    check = True
    j = 1

    while check:
      if p[i]**j > num:

        if j == 1:
          list2.append(j)
        else:
          list2.append(j-1)
        check = False
      j+=1
 
  return list2
```

```python
# Logic: Find all the prime numbers from 1-num, then find the largest exponents for each prime number that make it 
#        just smaller than num. Then find the product of the prime numbers raised to the exponents to find the smallest divisble number.

n = 20  #find the smallest number divisble by all numbers from 1-n
primes = [] #list to hold prime numbers
exp = []    #list to hold the relevant exponents
prd = 1
 
primes = find_prime(n)
exp = find_exp(n, primes)
print(primes)
print(exp)
 
for i in range(0,len(primes)):
  prd = prd*(primes[i]**exp[i])
 
print(prd)
```
